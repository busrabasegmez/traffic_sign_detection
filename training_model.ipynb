```python
from ultralytics import YOLO
from IPython.display import Image, display
import os
import cv2

# Load the trained model
model = YOLO('/content/drive/My Drive/YOLO_traffic_signs/runs/detect/traffic_signs_model2/weights/best.pt')

# Directory to save results
results_dir = '/content/drive/My Drive/YOLO_traffic_signs/test_results'
os.makedirs(results_dir, exist_ok=True)

# Path to test images directory
test_images_path = '/content/drive/My Drive/YOLO_traffic_signs/test_images'

# List all images in the test_images directory
test_images = [f for f in os.listdir(test_images_path) if f.endswith('.jpg') or f.endswith('.png') or f.endswith('.jpeg')]

# Run inference on each test image
for img_name in test_images:
    img_path = os.path.join(test_images_path, img_name)
    results = model(img_path)

    # Save the results
    for i, result in enumerate(results):
        # Save each image with annotations
        result_img_path = os.path.join(results_dir, f'{os.path.splitext(img_name)[0]}_result{i}.jpg')
        annotated_image = result.plot()
        cv2.imwrite(result_img_path, annotated_image)

        # Display the results
        display(Image(filename=result_img_path))
```
